## Train and Test Set

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:,:4]
y = df.iloc[:, 4]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state=42)
X_train.head(3)
```

```python
#stratify with target label y to have equal proportion and not have any skewdness in data.
print(y_train.value_counts()/len(y_train))
print(y_test.value_counts()/len(y_test))
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale.fit(X_train)
X_train_std = scale.transform(X_train)
X_test_std = scale.transform(X_test)
```

## Training the Models
<p>
    First we will be training using all the features that were given.
    </p>

```python
#First we are using the most basic logistic regression.
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

def logisticreg(X_train,y_train,X_test, y_test):
    model = LogisticRegression(multi_class = 'multinomial', solver  = 'lbfgs')
    model.fit(X_train, y_train )
    pred_y = model.predict(X_test)
    return metrics.accuracy_score(pred_y,y_test)

logisticreg(X_train_std,y_train,X_test_std, y_test)
```

```python
#using SVM classifier
from sklearn import svm

def SVM(X_train, y_train, X_test, y_test):
    model = svm.SVC(kernel = 'rbf', gamma='scale')#kernel = 'rbf', gamma='scale'
    model.fit(X_train, y_train)
    pred_y = model.predict(X_test)
    return metrics.accuracy_score(pred_y, y_test)

SVM(X_train_std,y_train,X_test_std, y_test)
```

<ul>
    <li>Using Logistic Regression we are able to get an accuracy of 93.33%.</li>
    <li>Using Support Vector Machine we are able to get an accuracy of 96.66%.</li>
    <li>In the above models we were using all the features, next we will try to seperate the train data in two Petal and Sepal and then try to fit and predict with our models with both of them.</li>
    </ul>

## Seperating the data in Sepal and Petal Features

```python
X_train_std_sep = X_train_std[:,:2]
X_test_std_sep = X_test_std[:, :2]

X_train_std_pet = X_train_std[:,2:]
X_test_std_pet = X_test_std[:, 2:]
```

```python
score_Sepal = logisticreg(X_train_std_sep, y_train,X_test_std_sep, y_test )
score_Petal = logisticreg(X_train_std_pet, y_train,X_test_std_pet, y_test )
print('Score for Sepal and Petal features with logistic regression models are ', score_Sepal, ' and ', score_Petal, '.')
```

```python
score_Sepal = SVM(X_train_std_sep, y_train,X_test_std_sep, y_test)
score_Petal = SVM(X_train_std_pet, y_train,X_test_std_pet, y_test)
print('Score for Sepal and Petal features with logistic regression models are ', score_Sepal, ' and ', score_Petal, '.')
```

```python
from matplotlib.colors import ListedColormap
from sklearn import preprocessing

def plot_decision_regions(X, y, classifier, name, test_idx=None, resolution=0.02):

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    #for idx, cl in enumerate(np.unique(y)):
    #    plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],alpha=0.8, c=cmap(idx),marker=markers[idx], label=cl)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='cubehelix', alpha=0.8)
    plt.xlabel('{} Length'.format(name))
    plt.ylabel('{} Width'.format(name))
    plt.title('{} Length Vs Width with normalize Data & Decision Boundary'.format(name))
    
def EncodeAndplotlogistic(X_train,y_train, X_test, y_test, name ):
    #encoding the label in 0, 1 , 2
    le = preprocessing.LabelEncoder()
    le.fit(y_test)
    y_train_new = le.transform(y_train)
    y_test_new = le.transform(y_test)
    
    model = LogisticRegression(multi_class = 'multinomial', solver  = 'lbfgs')
    model.fit(X_train, y_train_new)
    pred_y = model.predict(X_test)
    plot_decision_regions(X_test, y_test_new, model, name)
        

        
        
def EncodeAndplotsvm(X_train,y_train, X_test, y_test, name ):
    #encoding the label in 0, 1 , 2
    le = preprocessing.LabelEncoder()
    le.fit(y_test)
    y_train_new = le.transform(y_train)
    y_test_new = le.transform(y_test)
    
    model = svm.SVC(kernel = 'rbf', gamma='scale')#kernel = 'rbf', gamma='scale'
    model.fit(X_train, y_train_new)
    pred_y = model.predict(X_test)
    plot_decision_regions(X_test, y_test_new, model, name)
    
```

## Decision Boundaries with Logistic Regression

```python
EncodeAndplotlogistic(X_train_std_pet,y_train, X_test_std_pet, y_test, 'Petal' )
```

```python
EncodeAndplotlogistic(X_train_std_sep,y_train, X_test_std_sep, y_test, 'Sepal' )
```

## SVM Decsion Boundaries for Sepal And Petal datasets

```python
EncodeAndplotsvm(X_train_std_pet,y_train, X_test_std_pet, y_test, 'Petal' )
```

```python
EncodeAndplotsvm(X_train_std_sep,y_train, X_test_std_sep, y_test, 'Sepal' )
```
